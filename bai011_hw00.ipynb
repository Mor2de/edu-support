**Задание 6.** Постройте график функции, заданной параметрически: $\displaystyle x(t)=6 \sin{\left (8 t \right )}$, $\displaystyle y(t)=6 \sin{\left (3 t \right )}$. Используйте минимальный по длине промежуток изменения параметра и одинаковый масштаб по координатным осям.

```python
from sympy import *
init_printing()
%matplotlib inline
var('t')
plotting.plot_parametric(6*sin(8*t), 6*sin(3*t), (t, 0, 2*pi),aspect_ratio = (1, 1))
```
